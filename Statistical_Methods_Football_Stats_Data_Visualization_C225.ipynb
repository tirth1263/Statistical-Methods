```python
#C225_TIRTH_RANK

qh25, qh50, qh75 = np.percentile(data['m_played'], [25,50,75])
iqrh = qh75-qh25
minh = qh25 - 1.5*iqrh
maxh = qh75 + 1.5*iqrh

data = data[(data['m_played'] > minh) & (data['m_played'] < maxh)]
```

```python
#C225_TIRTH_RANK

qh25, qh50, qh75 = np.percentile(data['goals'], [25,50,75])
iqrh = qh75-qh25
minh = qh25 - 1.5*iqrh
maxh = qh75 + 1.5*iqrh

data = data[(data['goals'] > minh) & (data['goals'] < maxh)]
```

```python
#C225_TIRTH_RANK

qh25, qh50, qh75 = np.percentile(data['shots'], [25,50,75])
iqrh = qh75-qh25
minh = qh25 - 1.5*iqrh
maxh = qh75 + 1.5*iqrh

data = data[(data['shots'] > minh) & (data['shots'] < maxh)]
```
